10. How can you tell the difference between break and continue?

```python
for i in range(10):
  if (i==7):
    break
  print(i)

print("breaked")

for i in range(10):
  if(i==7):
    continue
  print(i)
```

ANS= Break is used for leaves the loop , skips remaining execution of complete loop and it is useful, if condition always evaluates to be True whereas continue is used for jumps for next iteration, skips execution of remaining statement inside the loop for current iteration and it is useful, if wants to skip execution of some statement nside the loop for any particular iteration.
